DATA REPORTING

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
```

```python
#load data from Data_cleaning_pandas notebook
countries=pd.read_csv('exports/countries_clean.csv', index_col=0)
```

```python
#get 10 most populated countries
most_populated=countries.sort_values('Population', ascending=False).head(50)
```

```python
#report cummulative sum for most populated countries
most_populated['Area (sq. mi.)'].cumsum()
```

```python
#get count of most populated coutries per region
most_populated['Region'].value_counts(sort=True)
```

```python
#group by Region and calculate population min, max sum and mean
most_populated.groupby('Region')['Population'].agg([min,max, sum, np.mean])
```

```python
#make pivot table with count values for regions and climate types
most_populated.pivot_table(values= 'Population', index='Region', columns='Climate', fill_value=0, margins= True, aggfunc='count')
```
